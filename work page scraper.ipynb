```python
def mov_scraper(link):
    try:
        pg = requests.get('https://elcinema.com'+link).text
        soup = bs(pg)
        box = soup.find('div', attrs={'class':'columns small-12 medium-9 large-9'})
        info1 = box.find_all('ul', attrs={'class':'list-separator'})
        zz = [[] for i in range(len(info1))]
        for i, item in enumerate(info1):
            for li in item.find_all('li'):
                zz[i].append(li)
        details = soup.find('ul', attrs={'id':'jump-here-genre'}).parent
        yy = {x.li.text.split()[0]: x('li')[1:] for x in details('ul', attrs={'class':'list-separator list-title'})}
        dd = inf(yy)
        try:
            dd['Director'] = [[li.a['href'], li.a.text] for li in zz[4][1:]]
        except:
            dd['Director'] = np.nan
        try:
            dd['Writer'] = [[li.a['href'], li.a.text] for li in zz[5][1:]]
        except:
            dd['Writer'] = np.nan
        try:
            cast = soup.find('div', attrs={'class':'boxed-0'}).find_all('ul', attrs={'class':'description'})
            cast = [[x.li.a.text, x.li.a['href'], x.li.next_sibling.next_sibling.text.strip()[1:-1]] for x in cast]
            dd['Cast'] = cast
        except:
            dd['Cast'] = np.nan
        try:
            dd['Story'] = details.p.text.strip()
        except:
            dd['Story'] = np.nan
        try:
            dd['Story'] = re.sub(r'\.\.\.اقرأ المزيد', '', dd['Story'])
        except:
            dd['Story'] = np.nan
        try:
            dd['Duration'] = int(zz[0][2].text.split()[0])
        except:
            dd['Duration'] = np.nan
        try:
            dd['Released'] = zz[0][3].text
        except:
            dd['Released'] = np.nan
        try:
            dd['Parts'] = [ [x.a.text, x.a['href']] for x in zz[7][1:]]
        except:
            dd['Parts'] = np.nan
        return dd
    except:
        return np.nan
```
